```python
rc_v_house = df_votes.merge(df_rollcalls_house, on='roll_call_id')

df_rc_v_house = df_votes[df_votes['roll_call_id'].isin(rc_v_house['roll_call_id'])]

df_rc_v_house.head()
```

```python
df_people_senate = df_people[df_people['role'] == 'Sen']

df_people_house = df_people[df_people['role'] == 'Rep']

print(df_people_senate.head())
print(df_people_house.head())
```
